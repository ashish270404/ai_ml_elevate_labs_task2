```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('G:\Titanic-Dataset.csv')
```

```python
df.describe() #obj1  >Generate summary statistics (mean, median, std, etc.).
```

```python
df.describe(include='object')
```

```python
# obj2 >for histograms of numberic columns 
for col in df.select_dtypes(include=np.number).columns:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()
```

```python
#obj2 >for boxplots
for col in df.select_dtypes(include=np.number).columns:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
#obj3 pairplot for feature relation
sns.pairplot(df)
plt.suptitle("Pairplot of Features", y=1.02)
plt.show()
```

```python
#obj3 corelation matrix
corr = df.corr(numeric_only=True)
plt.figure(figsize=(6, 4))
sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix")
plt.show()
```

```python
#obj4 surviaval rate by p class for different genders
sns.barplot(data=df, x='Pclass', y='Survived', hue='Sex')
plt.title('Survival Rate by Passenger Class and Gender')
plt.show()
```
